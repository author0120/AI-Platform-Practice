```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table('기생충review.txt', engine='python', encoding='cp949', header=None)
```

```python
df
```

```python
text = ''
for review in df[0]:
    text += review + ' '
```

```python
text
```

```python
#전처리 리뷰 분석 시 쓸데없는 문자, 부호들
special = ['.', '!', '♡', '~', ',', '?', '<', '>', ':', ';', '+', '-', '=', '"', '%', '(', ')', '^', '♥']
```

```python
nums = [] # 0 ~ 9 까지
for i in range(10):
    nums.append(str(i))
```

```python
nums
```

```python
for x in (special + nums):
    text = text.replace(x, '')
```

```python
text
```

# 정규표현식

```python
import re
```

```python
#대문자 A부터 Z까지, 소문자 a부터 z까지 / 모든 영문자
# ^ -> 시작 즉, 모든 영문자로 시작하면
# 특수문자 하나하나 입력 안 해도 한번에 없앨 수 있다.
# 불필요한것 제거하는 것 -> 전처리
text = re.sub('[^A-Za-z가-힣\\s]', '', text)
```

```python
text
```

```python
# w -> 읽기모드
with open('기생충review_최종.txt', 'w', encoding = 'euc-kr') as f:
    f.write(text)
```

```python
with open('기생충dict.txt', encoding = 'euc-kr') as f:
    terms = f.read().split('\n')[:-1]  #끝까지 읽기
```

```python
terms
```

```python
from collections import Counter
from konlpy.tag import Okt
```

# https://github.com/ojdkbuild/ojdkbuild

# java-1.8.0-openjdk-1.8.0.332-1.b09.ojdkbuild.windows.x86_64.zip (sha256)  다운로드

```python
nlp = Okt()
```

```python
nouns = nlp.nouns(text)
count_dict = Counter(nouns)
```

```python
count_dict
```

```python
freq = []
for term in terms:
    freq.append(count_dict[term])
```

```python
freq
```

```python
tdm = pd.DataFrame({'term':terms, 'freq':freq}).set_index('term').sort_values(by='freq', ascending=False)
```

```python
tdm
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.family'] = 'gulim'
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(1, 1)
plt.title('기생충 리뷰의 사전 빈도 시각화')
tdm.plot.bar(ax=ax, rot=45)
plt.ylabel('빈도')
plt.show()
```
